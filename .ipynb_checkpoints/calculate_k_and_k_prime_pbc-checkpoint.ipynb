```python
df_rf = pd.read_csv("__hbrf/124_2LiI_PBC_ns10_rfachb_h_17.csv", names= ["time","rf"])
df_3 = df_rf.rf             
```

```python
df_data = pd.concat([df_c, df_3, df_n],axis=1 )
df_data[0:40]
```

说明：

rf: $k(t)$;

n: $n(t)$;

c: $c(t)$, 又记为$C_{HB}(t)$.

已知 $k(t),n(t),c(t)$, 假设它们满足关系式

$$k(t) = kc(t)-k'n(t),$$
其中，$k,k'$是两个常量. (注： $k$,和$k(t)$是不一样的.)

下面要做的是，计算$\Delta=[kc(t)-k'n(t)-k(t)]^2 $. 找出能使得$\Delta$的值最小的$k$和$k'$.

```python
#用最小二乘法拟合出k和k'
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
#plt.rcParams['font.sans-serif']=['SimHei'] # 用来正常显示中文标签
#plt.rcParams['axes.unicode_minus']=False # 用来正常显示负号
from matplotlib import font_manager as fm, rcParams

fig = plt.figure()
ax = fig.gca(projection='3d')

c = np.array(df_data.c)
n = np.array(df_data.n)
rf = np.array(df_data.rf)

# Make data.
delta_x = 0.01
delta_y = 0.01
X = np.arange(0.0, 1, delta_x)
Y = np.arange(0.0, 1, delta_y)
X, Y = np.meshgrid(X, Y)
R = np.zeros((100,100))
for i in range(61,2400):
    R += (X*c[i] - Y*n[i] -rf[i])**2
    
Z = R
 
# Plot the surface.
#surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False)
surf = ax.plot_surface(X, Y, Z, cmap=cm.plasma, linewidth=0, antialiased=False) 

# Customize the z axis.
ax.set_zlim(0.0, 1000.0)
ax.zaxis.set_major_locator(LinearLocator(5))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.1f'))

# Add a color bar which maps values to colors.
#  The unit of k and k' is: (0.1 ps^{-1}).
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_xlabel(r'$k $') 
ax.set_ylabel(r"$k'$")
ax.set_zlabel(r"$ [kC_{HB}(t)-k'n(t)-k(t)]^2 $")
plt.show()

#print(R) 
for i in range(0,100,1):
    print ("i =", i)
    for j in range(0,100,1):
        print(i*delta_x, j*delta_y, R[i][j])
```
