```python
import numpy as np
import pandas as pd
```

```python
df_c = pd.read_csv("_hbacf_continuous/124_2LiI_hbacf_h.csv", names= ["time","c"])
df_c
```

```python
df_rf_in_and_n = pd.read_csv("__hb_reaction_flux/124_2LiI_rfhb_in_h--n.csv", names= ["time","rf_in","n"])
df_n = df_rf_in_and_n.n
df_n
```

```python
df_rf = pd.read_csv("__hbrf/124_2LiI_rfachb_h_17.csv", names= ["time","rf"])
df_3 = df_rf.rf
```

```python
df_data = pd.concat([df_c, df_3,df_n],axis=1 )
df_data
```

```python
delta = 0.01
c = df_data.c
n = df_data.n
rf = df_data.rf
delta2 =[]
para = []
for m1 in range(100):
    k=delta*m1
    for m2 in range(100):
        k_prime=delta*m2
        rf_temp = k*c - k_prime * n
        delta2.append(np.sum((rf_temp - rf)**2))
        para.append((k,k_prime))
```
